### Assessment
#### Labeling variables and values of all categorical values

```python
# Task 1
# Labeling all variables in the main and roster datasets
# For the main dataset;

main_df.columns
```

```python
new_column_names = {'province_id': 'Province code',
                    'province_name': 'Province of residence',
                    'district_id': 'District code',
                    'district_name': 'District of residence',
                    'sector_id': 'Sector code',
                    'sector_name': 'Sector of residence',
                    'cell_id': 'Cell code',
                    'cell_name': 'Cell of residence',
                    'village_id': 'Village code',
                    'village_name': 'Village of residence',
                    'list_available': 'If the ubudehe list is available',
                    'list_source': 'From whom the ubudehe list was obtained',
                    'ubudehe_contact': 'Ubudehe list contact person',
                    'contact_role': 'Role of the ubudehe list contact person',
                    'contact_phone': 'Phone number of ubudehe list contact person',
                    'num_households': 'Number of households in ubudehe category 1',
                    'parent_key': 'Unique ID'
                    }
main_df.rename(columns=new_column_names, inplace = True)
```

```python
main_df.sample(3)
```

```python
# creating a new column named gps 
main_df['gps'] = main_df['gpslatitude'].astype(str) + ', ' + main_df['gpslongitude'].astype(str)
# dropping unnecessary columns
columns_dropped = ['gpslatitude', 'gpslongitude', 'gpsaltitude', 'gpsaccuracy']
main_df = main_df.drop(columns=columns_dropped)
main_df.head(3)
```

```python
# For the roster dataset
roster_df.columns
```

```python
New_roster_columns = { 'hh_head_position': 'Position of household head on ubudehe list',
                        'name_hh_head': 'Name of the household head',
                        'nid_hh_head': 'National ID number of the household head',
                        'phone_hh_head': 'Phone number of household head',
                        'parent_key': 'Unique ID'
                     }
roster_df.rename(columns=New_roster_columns, inplace = True)
roster_df.head(3)
```

```python
# label the values of all categorical variables in both the main and the roster datasets
# The columns with categorical values are 'If the ubudehe list is available', 
#'From whom the ubudehe list was obtained',  'Role of the ubudehe list contact person'. All from the main dataset
value_labels = { "If the ubudehe list is available": {1: "Yes", 0: "No"}, 
                 "From whom the ubudehe list was obtained": {1: "Sector Executive Officer", 2: "Cell Executive Officer", 3: "Village Leader"},
                 "Role of the ubudehe list contact person": {1: "Village leader", 2: "Village in-charge of security", 3: "Village Community Health Worker"}
               }
for x, y in value_labels.items():
    main_df[x] = main_df[x].map(y)
```

```python
# value counts for 'If the ubudehe list is available'
print("\nValue counts for 'If the ubudehe list is available':")
print(main_df['If the ubudehe list is available'].value_counts())

# Print value counts for 'From whom the ubudehe list was obtained'
print("\nValue counts for 'From whom the ubudehe list was obtained':")
print(main_df['From whom the ubudehe list was obtained'].value_counts())

# Print value counts for 'Role of the ubudehe list contact person'
print("\nValue counts for 'Role of the ubudehe list contact person':")
print(main_df['Role of the ubudehe list contact person'].value_counts())
```

#### Creating a single variable for each of the location variables (province, district, sector, cell, and village)

```python
# Task 2
# create a single variable for each of the location variables (province, district, sector, cell, and village)
columns_to_concat = ['Province', 'District', 'Sector', 'Cell', 'Village']
for c in columns_to_concat:
    id_col = f"{c} code"
    name_col = f"{c} of residence"
    main_df[c] = main_df[[id_col, name_col]].apply(lambda row: f"{row[id_col]}_{row[name_col]}", axis=1)
```

```python
# We need to drop unnecessary columns after concatanating columns above
columns_to_drop = ['Province code', 'Province of residence', 'District code', 'District of residence', 
                   'Sector code','Sector of residence', 'Cell code', 'Cell of residence', 'Village code', 
                   'Village of residence']
main_df = main_df.drop(columns=columns_to_drop)
main_df.columns
```

```python
# Checking the concatanated columns' values
main_df[['Province', 'District', 'Sector', 'Cell', 'Village']].head(10)
```

#### Merging the datasets

```python
# Task 3
# Merge both datasets so that we have one observation per village with all household heads as well as their personal information contained in the roster.
merged_93302_20240101 = pd.merge(main_df, roster_df, on= 'Unique ID', how='inner')
merged_93302_20240101.head(3)
```

#### Extraction
Extract lists of households per district in .xlsx format, with each sector in each of the 3 districts on
a separate worksheet.


```python
# Task 5
# Printing out unique values from the 'district' column in the dataset
clean_93302_20240101['District'].unique()
```

```python
# districts = ['12_Gasabo', '11_Nyarugenge', '13_Kicukiro']
# Checking the number or records each district has below
```

```python
gasabo_df = clean_93302_20240101[['District', 'Sector']][clean_93302_20240101['District'] == '12_Gasabo']
gasabo_df.shape
```

```python
nyarugenge_df = clean_93302_20240101[['District', 'Sector']][clean_93302_20240101['District'] == '11_Nyarugenge']
nyarugenge_df.shape
```

```python
kicukiro_df = clean_93302_20240101[['District', 'Sector']][clean_93302_20240101['District'] == '13_Kicukiro']
kicukiro_df.shape
```

```python
# Display the DataFrames
print("Filtered Data for Gasabo District:\n", gasabo_df.sample(5), "\n" + "="*50 + "\n")
print("Filtered Data for Nyarugenge District:\n", nyarugenge_df.sample(5), "\n" + "="*50 + "\n")
print("Filtered Data for Kicukiro District:\n", kicukiro_df.sample(5), "\n" + "="*50 + "\n")
```

```python
# Extract the documents
with pd.ExcelWriter('93302_20240101.xlsx', engine='xlsxwriter') as writer:
    # Write each DataFrame to a separate worksheet
    gasabo_df.to_excel(writer, sheet_name='Gasabo', index=False)
    nyarugenge_df.to_excel(writer, sheet_name='Nyarugenge', index=False)
    kicukiro_df.to_excel(writer, sheet_name='Kicukiro', index=False)
```
